```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv("drive/My Drive/Colab Notebooks/winequality-white.csv", delimiter=',')
data
```

```
data.corr()
```

```
data.cov()
```

```
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
```

```
gaussian_model = GaussianNB()
bernoulli_model = BernoulliNB()
multinomial_model = MultinomialNB()
```

```
values = data.values
X = values[:,:-1]
Y = values[:,-1]

gaussian_model.fit(X,Y)
bernoulli_model.fit(X,Y)
multinomial_model.fit(X,Y)
```

```
print(gaussian_model.score(X,Y),
      bernoulli_model.score(X,Y),
      multinomial_model.score(X,Y))
```
